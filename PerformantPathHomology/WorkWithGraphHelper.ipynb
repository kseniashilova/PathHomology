```python
import numpy as np
import pandas as pd
```

## Fly

```python
data = pd.read_csv('fly.txt', delimiter=' ', names=['s', 't', 'w'])
data
```

```python
#Получить уникальные номера вершинок
```

```python
def is_exist(lst, i):
    for el in lst:
        if i == el:
            return True
    return False
```

```python
lst = []
for i in data['s']:
    if not is_exist(lst, i):
        lst.append(i)
```

```python
for i in data['t']:
    if not is_exist(lst, i):
        lst.append(i)
```

```python
len(lst)
```

```python
data['s_fix'] = data['s'].apply(lambda x: lst.index(x) + 1)
```

```python
data['t_fix'] = data['t'].apply(lambda x: lst.index(x) + 1)
```

```python
data
```

```python
new_list = [str(i)+'\n' for i in data['s_fix'].to_list()]
with open('fly_start_neurons.txt', 'w') as f:
          f.writelines(new_list)
```

```python
new_list = [str(i)+'\n' for i in data['t_fix'].to_list()]
with open('fly_end_neurons.txt', 'w') as f:
          f.writelines(new_list)
```

## C elegans

```python
data = pd.read_csv('celegans.txt', delimiter=' ', names=['s', 't', 'w'])
data
```

```python
data = data[data['s'] != data['t']]
data
```

```python
lst = []
for i in data['s']:
    if not is_exist(lst, i):
        lst.append(i)
for i in data['t']:
    if not is_exist(lst, i):
        lst.append(i)
```

```python
len(lst)
```

```python
data['s_fix'] = data['s'].apply(lambda x: lst.index(x) + 1)
data['t_fix'] = data['t'].apply(lambda x: lst.index(x) + 1)
```

```python
data
```

```python
new_list = [str(i)+'\n' for i in data['s_fix'].to_list()]
with open('celegans_start_neurons.txt', 'w') as f:
          f.writelines(new_list)
```

```python
new_list = [str(i)+'\n' for i in data['t_fix'].to_list()]
with open('celegans_end_neurons.txt', 'w') as f:
          f.writelines(new_list)
```
