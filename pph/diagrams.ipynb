### Getting intervals

```python
def get_integer_intervals_from_file(file_name, dim):
    intervals = []
    with open(file_name) as f:
        intervals = f.read().split('are')
    #intervals[1] - for dim = 0
    #intervals[2] - for dim = 1
    
    #удаляем последний элемент
    new_intervals = (intervals[dim + 1].split('(')[1:len(intervals[dim + 1].split('(')[1:])])
    #добавляем последний элемент в нужном формате
    last_el = (intervals[dim + 1].split('(')[1:][len(intervals[dim + 1].split('(')[1:]) - 1]).split(']]')[0]
    new_intervals.append(last_el)
    
    integer_intervals = []
    for interval in new_intervals:
        integer_intervals.append(
            [(int)(interval.split(')')[0].split(',')[0].split('.')[0]),
             (int)(interval.split(')')[0].split(',')[1].split('.')[0])])
    return integer_intervals
```

```python
celegans_dim0 = np.array(get_integer_intervals_from_file('celegans_intervals.txt', 0))
celegans_dim1 = np.array(get_integer_intervals_from_file('celegans_intervals.txt', 1))
```

```python
fly_dim0 = np.array(get_integer_intervals_from_file('fly_intervals.txt', 0))
fly_dim1 = np.array(get_integer_intervals_from_file('fly_intervals.txt', 1))
```

#### Betti curves

```python
def get_betti_numbers(intervals):
    betti = []
    for i in range(0,10001):
        sum = 0
        for interval in intervals:
            if (interval[0] <= i) and (interval[1] > i):
                sum = sum + 1
        betti.append(sum)
    return betti
```

```python
plt.plot(range(0,10001), get_betti_numbers(celegans_dim0), 'yo')
```

```python
plt.plot(range(0,10001), get_betti_numbers(celegans_dim1), 'yo')
```

```python
plt.plot(range(0,10001), get_betti_numbers(fly_dim0), 'yo')
```

```python
plt.plot(range(0,10001), get_betti_numbers(fly_dim1), 'yo')
```
